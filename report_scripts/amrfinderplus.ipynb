```python
import pandas as pd
```

```python
samples = [
    "cc4_1", "cc4_2", "cc4_3",
    "cd4_1", "cd4_2", "cd4_3",
    "mc4_1", "mc4_2", "mc4_3",
    "md4_1", "md4_2", "md4_3",
    "fc4_1", "fc4_2", "fc4_3",
    "fd4_1", "fd4_2", "fd4_3",
]
```

```python
for s in samples:
    print(s)
    samplefile = f"../results_metaspades/{s}/amr_finderplus/log.out"
    df = pd.read_csv(samplefile, sep="\t")
    print(df.iloc[:, [1,2,3]])
    print("============")
    
```

```python

samples_df = {}

all_gene_symbol = []

for s in samples:
#     print(s)
    samplefile = f"../results_metaspades/{s}/amr_finderplus/log.out"
    df = pd.read_csv(samplefile, sep="\t")
    df.rename(columns={"Gene symbol": "gene_symbol"}, inplace=True)
    
    all_gene_symbol += list(df.gene_symbol.unique())
    grp = df.groupby(by=["gene_symbol"]).count().reset_index().loc[:,["gene_symbol", "Contig id"]]
    
    samples_df[s] = grp
    

    
    
unique_gene_symbol = set(all_gene_symbol)

m = pd.DataFrame({"gene_symbol": list(unique_gene_symbol)})
m

for s, sdf in samples_df.items():
    sdf.rename(columns={"Contig id": s}, inplace=True)
    m = pd.merge(m, sdf, how="left", on="gene_symbol")
    
```
