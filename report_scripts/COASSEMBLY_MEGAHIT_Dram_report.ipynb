```python
def get_taxa_ranks(classification, taxa_rank="all"):
    mapping = {
        "domain": 'd',
        "phylum": 'p',
        "class": 'c',
        "order": 'o',
        "family": 'f',
        "genus": 'g',
        "species": 's',
    }
    
    tokens = classification.split(";")
    d = {}
    for tkn in tokens:
        rank = tkn[0]
        info = tkn[3:]
        
        d[rank] = info
    
    if taxa_rank == "all":
        return d
    elif taxa_rank == "domain":
        return d[mapping[taxa_rank]]
    elif taxa_rank == "phylum":
        return d[mapping[taxa_rank]]
    elif taxa_rank == "class":
        return d[mapping[taxa_rank]]
    elif taxa_rank == "order":
        return d[mapping[taxa_rank]]
    elif taxa_rank == "family":
        return d[mapping[taxa_rank]]
    elif taxa_rank == "genus":
        return d[mapping[taxa_rank]]
    else:
        assert taxa_rank == "species"
        return d[mapping[taxa_rank]]
        
```

```python
m["domain"]  = m.apply(lambda x: get_taxa_ranks(x["Classification"], "domain"), axis=1)
m["phylum"]  = m.apply(lambda x: get_taxa_ranks(x["Classification"], "phylum"), axis=1)
m["class"]   = m.apply(lambda x: get_taxa_ranks(x["Classification"], "class"), axis=1)
m["order"]   = m.apply(lambda x: get_taxa_ranks(x["Classification"], "order"), axis=1)
m["family"]  = m.apply(lambda x: get_taxa_ranks(x["Classification"], "family"), axis=1)
m["genus"]   = m.apply(lambda x: get_taxa_ranks(x["Classification"], "genus"), axis=1)
m["species"] = m.apply(lambda x: get_taxa_ranks(x["Classification"], "species"), axis=1)
```
